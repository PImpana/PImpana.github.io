```python
import pandas as pd

# Import an Excel or CSV file
# Replace 'data.csv' or 'data.xlsx' with your actual file path
file_path = 'https://raw.githubusercontent.com/PImpana/PImpana.github.io/refs/heads/main/Cc10/Borough%26typeofcrime.csv'  # For CSV


# Load the data
if file_path.endswith('.csv'):
    df = pd.read_csv(file_path)
else:
    raise ValueError("Unsupported file type. Use CSV or Excel.")

# Display the original wide format
print("Wide Format:")
print(df)

# Convert from wide to long format
# 'id_vars' are the columns you want to keep fixed (e.g., Borough, Crime Type)
# 'var_name' is the new column name for the wide-form column headers
# 'value_name' is the new column name for the data in wide-form cells
long_df = pd.melt(df,
                  id_vars=['Borough'],  # Change this based on your data structure
                  var_name='Crime Type',
                  value_name='Crime')

# Display the converted long format
print("\nLong Format:")
print(long_df)

# Save the long-form data to a new file
long_df.to_csv('boroughcrimetype_long_format.csv', index=False)  # Save as CSV
# long_df.to_excel('long_format.xlsx', index=False)  # Save as Excel
```
